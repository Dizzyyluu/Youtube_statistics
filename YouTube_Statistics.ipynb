Import the libraries for the analysis and visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#Reading the dataset into a DataFrame
df = pd.read_csv("Global Youtube Statistics.csv", encoding = "latin1")
```

```python
#Get descriptive statistics of numerical columns on the DataFrame
df.describe()
```

```python
# Checking for missing values in the DataFrame
df.isna().sum()
```

```python
# Checking for duplicate rows in the DataFrame
df.duplicated().sum()
```

```python
#Getting the column names of the DataFrame
df.columns
```

Analyzing the mean number of subscribers by category and visualizing the top 7 categories with the highest mean subscribers using a bar plot

```python
#Calculating the mean number of subscribers by category
category_subscribers_mean = df.groupby("category")["subscribers"].mean()
```

```python
#Formats the floating-point numbers to display with two decimal places
pd.set_option("display.float_format", "{:.2f}".format)
```

```python
category_subscribers_mean
```

```python
#Sorting the mean subscribers by category in descending order
```

```python
category_subscribers_mean.sort_values(ascending = False, inplace = True)
```

```python
# Plotting the top 7 categories by mean subscribers
```

```python
category_subscribers_mean.head(7).plot(kind = "bar")
plt.title("Mean Subscribers by Category")
plt.xlabel("Category")
plt.ylabel("Mean Subscribers")
plt.show()
```

Create a scatter plot to explore the relationship between subscribers and video views.

```python
plt.scatter(df["subscribers"], df["video views"])
plt.title("Scatter Plot of Subscriber vs Video Views")
plt.xlabel("Subscribers")
plt.ylabel("Video Views")
plt.show()
```

Calculate the total number of subscribers by country and visualizes the top 10 countries with the highest total subscribers using a bar plot.

```python
#Calculate the total number of subscribers by country and sort them in descending order
country_subs_total = df.groupby("Country")["subscribers"].sum().sort_values(ascending = False)
```

```python
country_subs_total
```

```python
country_subs_total.head(10).plot(kind = "bar")
plt.title("Total Subscribers by Country")
plt.xlabel("Country")
plt.ylabel("Total Subscribers")
plt.xticks(rotation = 90)
plt.show()
```

Visualize the distribution of the highest monthly earnings for different channel types using a box plot.

```python
sns.boxplot(x = "channel_type", y ="highest_monthly_earnings", data =df)
plt.title("Distrib. of Highest Monthly Earning by Channel Type")
plt.xticks(rotation = 90)
plt.show()
```

Creating a stacked bar plot to visualize the breakdown of subscribers by category and country

```python
df.groupby(["category", "Country"])["subscribers"].sum().unstack().plot(kind = "bar", stacked = True)
plt.title("Subscribers Breakdown by Category and Country")
plt.xlabel("Ctaegory")
plt.ylabel("Total Subscribers")
plt.legend(title = "Country", bbox_to_anchor = (1.05, 1), loc = "upper left")
plt.xticks(rotation = 90)
plt.show()
```

Creating a scatter plot to explore the relationship between subscribers vs gross tertiary education enrollment

```python
plt.scatter(df["Gross tertiary education enrollment (%)"], df["subscribers"], alpha = 0.5)
plt.title("Scatter plot of subscribers vs gross tertiary education enrollment")
plt.xlabel("Gross tertiary education enrollment")
plt.ylabel("Subscribers")
plt.show()
```

Creating a bar plot of average unemployment rate by channel type

```python
df.groupby("channel_type")["Unemployment rate"].mean().head(6).sort_values(ascending = False).plot(kind = "bar")
plt.title("Average Unemployment rate")
plt.xlabel("Channel Type")
plt.ylabel("Average Unemployment rate")
plt.xticks(rotation = 90)
plt.show()
```

Creating a bar plot of average subscribers by created year and channel type

```python
df.groupby(["created_year", "channel_type"])["subscribers"].mean().unstack().tail().plot(kind = "bar", width = 1)
plt.title("Average Subscribers by Created Year and Channel Type")
plt.xlabel("Created Year")
plt.ylabel("Average_Subscribers")
plt.xticks(rotation = 45)
plt.legend(title = "Channel_type", bbox_to_anchor =(1.05, 1), loc = "upper left")
plt.show()
```

Creating an area plot of subscribers growth over time, grouped by year and month

```python
#stacked area chart
df.groupby(["created_year", "created_month"])["subscribers"].sum().unstack().head().plot(kind ="area", stacked =True)
plt.title("Subscribers Growth over time")
plt.xlabel("Year")
plt.ylabel("Total Subscribers")
plt.legend(title = "Month", bbox_to_anchor = (1.05, 1), loc = "upper left")
plt.xticks(rotation = 90)
plt.show()
```

Creating a line plot of the trend of highest yearly earnings by created year and country

```python
df.groupby(["created_year", "Country"])["highest_yearly_earnings"].mean().unstack().head().plot(kind = "line", marker ="o")
plt.title("Trend of Highest Yearly Earnings by Created Year and Country")
plt.xlabel("Created Year")
plt.ylabel("Average highest yearly earnings")
plt.legend(title = "Country", bbox_to_anchor = (1.05, 1), loc = "upper left")
plt.xticks(rotation = 45)
plt.show()
```

Creating a heatmap to visualize the mean of subscribers for the last 30 days by channel type and created year

```python
sns.heatmap(channel_created_subscribers_mean.head(9), cmap = "YlGnBu", annot = True)
plt.title("Mean of Subscribers for the last 30 days by Channel Type and Created Year")
plt.xlabel("Created Year")
plt.ylabel("Channel Type")
plt.show()
```
